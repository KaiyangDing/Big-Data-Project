```python
from pyspark.sql import SparkSession

spark = SparkSession.builder \
    .appName("MyAnalysis") \
    .master("spark://spark-master:7077") \
    .config("spark.driver.memory", "4g") \
    .getOrCreate()

spark.range(10).show()
```

```python
# 读取航班数据
flights_df = spark.read.parquet("/data/processed/flights_clean")

print(f"航班记录数: {flights_df.count():,}")
print(f"字段数: {len(flights_df.columns)}")
print("\n前5行:")
flights_df.show(5, truncate=False)
```

```python
# 读取天气数据
weather_df = spark.read.parquet("/data/processed/weather_clean")

print(f"天气记录数: {weather_df.count():,}")
print(f"字段数: {len(weather_df.columns)}")
print("\n前5行:")
weather_df.show(5, truncate=False)
```
